```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from PIL import Image
import os
```

## Função para ler e juntar os CSV

```python
def load_datasets(folder):

    data_312 = []
    data_41 = []

    for file in os.listdir(folder):
        if file.endswith(".csv"):
            path = os.path.join(folder, file)

            df = pd.read_csv(path)

            if file.endswith("312.csv"):
                data_312.append(df)

            elif file.endswith("41.csv"):
                data_41.append(df)

    df_312 = pd.concat(data_312, ignore_index=True)
    df_41 = pd.concat(data_41, ignore_index=True)

    return df_312, df_41
```

## Função para adicionar Logos

```python
def add_logos(ax, languages, x_positions):

    logo_paths = {
        "C": "images/C_Logo.png",
        "Python": "images/Python_Logo.png",
        "JavaScript": "images/JavaScript_Logo.png",
        "Haskell": "images/Haskell_Logo.png"
    }

    target_height = 405

    for x, lang in zip(x_positions, languages):

        img = Image.open(logo_paths[lang]).convert("RGBA")
        ratio = target_height / img.height
        new_width = int(img.width * ratio)
        img = img.resize((new_width, target_height))

        imagebox = OffsetImage(img, zoom=0.08)
        ab = AnnotationBbox(
            imagebox,
            (x, 0),
            xybox=(0, -35),
            frameon=False,
            xycoords=('data', 'axes fraction'),
            boxcoords="offset points"
        )

        ax.add_artist(ab)
```

## Criação dos DataSets

```python
# Import csv data sets

os.makedirs('graphics', exist_ok=True)

base_path = "../StandardMeasurements"

powercap_path = os.path.join(base_path, "With Powercap")
nopowercap_path = os.path.join(base_path, "No Powercap")
```

```python
powercap312, powercap41 = load_datasets(powercap_path)
nopowercap312, nopowercap41 = load_datasets(nopowercap_path)
```

## Função para calcular as médias por Linguagem

```python
def compute_means(df312, df41, column):

    mean312 = df312.groupby("Language")[column].mean()
    mean41 = df41.groupby("Language")[column].mean()

    result = pd.DataFrame({
        "312": mean312,
        "41": mean41
    })

    return result
```

```python
def compute_means_by_input(df_powercap, df_nopowercap, column):
    mean_powercap = df_powercap.groupby("Language")[column].mean()
    mean_nopowercap = df_nopowercap.groupby("Language")[column].mean()

    result = pd.DataFrame({
        "With Powercap": mean_powercap,
        "No Powercap": mean_nopowercap
    })

    return result
```

## Criar gráficos de Barras

```python
def plot_bar(data, title, ylabel, filename, ymax=None):

    languages = data.index
    x = range(len(languages))

    width = 0.35

    plt.figure()

    plt.bar([i - width/2 for i in x], data["312"], width, label="Input (3,12)", color="#ff914d")
    plt.bar([i + width/2 for i in x], data["41"], width, label="Input (4,1)", color="#4c8dcf")

    plt.xticks(x, [])

    plt.ylabel(ylabel)
    plt.title(title)

    if ymax:
        plt.ylim(0, ymax)

    plt.legend(loc='upper left')
    # plt.grid(axis="y", linestyle="--", alpha=0.7)

    ax = plt.gca()

    add_logos(ax, languages, x)

    plt.subplots_adjust(bottom=0.2)

    plt.savefig(f"graphics/{filename}.png", dpi=300, bbox_inches="tight")
    plt.show()
```

```python
def plot_bar_input(data, title, ylabel, filename, ymax=None):

    languages = data.index
    x = range(len(languages))

    width = 0.35

    plt.figure()

    plt.bar([i - width/2 for i in x], data["With Powercap"], width, label="Powercap = 7", color="#ff914d")
    plt.bar([i + width/2 for i in x], data["No Powercap"], width, label="No Powercap", color="#4c8dcf")

    plt.xticks(x, [])

    plt.ylabel(ylabel)
    plt.title(title)

    if ymax:
        plt.ylim(0, ymax)

    plt.legend(loc='upper left')
    # plt.grid(axis="y", linestyle="--", alpha=0.7)

    ax = plt.gca()

    add_logos(ax, languages, x)

    plt.subplots_adjust(bottom=0.2)

    plt.savefig(f"graphics/{filename}.png", dpi=300, bbox_inches="tight")
    plt.show()
```

## Calcular média de Package

```python
energy_312 = compute_means_by_input(powercap312, nopowercap312, "Package")
energy_41 = compute_means_by_input(powercap41, nopowercap41, "Package")

# Calcular o eixo do y
ymax_energy = max(
    energy_312.values.max(),
    energy_41.values.max()
)

# Adicionar margem para não ficar colado ao teto
ymax_energy *= 1.1
```

## Energy Consumption - Input (3,12)

```python
plot_bar_input(
    energy_312,
    "Energy Consumption - Input (3,12)",
    "Package",
    "StandardMeasurements/energy_3_12",
    ymax_energy
)
```

## Energy Consumption - Input (4,1)

```python
plot_bar_input(
    energy_41,
    "Energy Consumption - Input (4,1)",
    "Package",
    "StandardMeasurements/energy_4_1",
    ymax_energy
)
```

## Package - Powercap vs No Powercap

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))

dir_std = "graphics/StandardMeasurements/"

img1 = mpimg.imread(dir_std + "energy_3_12.png")
img2 = mpimg.imread(dir_std + "energy_4_1.png")

ax[0].imshow(img1)
ax[0].axis("off")
# ax[0].set_title("Powercap")

ax[1].imshow(img2)
ax[1].axis("off")
# ax[1].set_title("No Powercap")

plt.savefig(f"{dir_std}energy_comparision.png", dpi=300, bbox_inches="tight")

plt.show()
```

## Calcular média de Tempo de Execução

```python
time_312 = compute_means_by_input(powercap312, nopowercap312, "Time")
time_41 = compute_means_by_input(powercap41, nopowercap41, "Time")

ymax_time = max(
    time_312.values.max(),
    time_41.values.max()
)

# Adicionar margem para não ficar colado ao teto
ymax_time *= 1.1
```

## Tempo de Execução - Input (3,12)

```python
plot_bar_input(
    time_312,
    "Execution Time - Input (3,12)",
    "Time (ms)",
    "StandardMeasurements/time_3_12",
    ymax_time
)
```

## Tempo de Execução - Input (4,1)

```python
plot_bar_input(
    time_41,
    "Execution Time - Input (4,1)",
    "Time (ms)",
    "StandardMeasurements/time_4_1",
    ymax_time
)
```
