```python
import os
import pandas as pd
import re
```

```python
folder = "../ValgrindGraphics/OutFiles"
```

```python
def parse_valgrind_file(filepath):
    data = []

    snapshot = None
    time = None
    mem_heap = None
    mem_heap_extra = None
    mem_stacks = None

    with open(filepath, "r") as f:
        for line in f:
            line = line.strip()

            if line.startswith("snapshot="):
                snapshot = int(line.split("=")[1])

            elif line.startswith("time="):
                time = int(line.split("=")[1])

            elif line.startswith("mem_heap_B="):
                mem_heap = int(line.split("=")[1])

            elif line.startswith("mem_heap_extra_B="):
                mem_heap_extra = int(line.split("=")[1])

            elif line.startswith("mem_stacks_B="):
                mem_stacks = int(line.split("=")[1])

                # quando chegamos aqui temos todos os valores
                data.append({
                    "snapshot": snapshot,
                    "time": time,
                    "mem_heap_B": mem_heap,
                    "mem_heap_extra_B": mem_heap_extra,
                    "mem_stacks_B": mem_stacks
                })

    return pd.DataFrame(data)
```

```python
for filename in os.listdir(folder):

    filepath = os.path.join(folder, filename)

    if os.path.isfile(filepath):

        df = parse_valgrind_file(filepath)

        csv_name = os.path.splitext(filename)[0] + ".csv"
        csv_path = os.path.join(folder, csv_name)

        df.to_csv(csv_path, index=False)

        print(f"Convertido: {filename} -> {csv_name}")
```
